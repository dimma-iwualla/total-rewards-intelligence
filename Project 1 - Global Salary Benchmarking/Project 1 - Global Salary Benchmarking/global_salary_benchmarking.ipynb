```python
import pandas as pd

df = pd.read_csv("ds_salaries.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby("job_title")["salary_in_usd"].median().sort_values(ascending=False)
```

```python
df.groupby("job_title")["salary_in_usd"].agg(["median", "count"]).sort_values(by="median", ascending=False)
```

```python
df_filtered = df.groupby("job_title")["salary_in_usd"].agg(["median", "count"])
df_filtered = df_filtered[df_filtered["count"] >= 10]
df_filtered.sort_values(by="median", ascending=False)
```
